```python
%matplotlib inline

from shutil import copyfile
from time import time
from collections import Counter
# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
plt.interactive(False)
print(tf.__version__)


print("imported...")

'''
This  program reads the Label.xls file  and copies the images belonging to different classes into respective folders

for ex: label: 0 0 1 0 0 --> /minor/
label: 0 1 0 0 0 --> /good/
label: 0 0 0 1 0 --> /major/
label: 0 0 0 0 1 --> /other/


Note: 1.  it reads the images from ./imaage_data/
2. Please make sure that you have already created the class folders(ie. ./major/ , ./minor/ ......) ''
3. After the copying is done, use the cropping python program(CropImage2.py) to crop the images.  manually.

'''
```

```python
df=pd.read_excel('Lables.xlsx')
print("file loaded")

fan= df["Fan"]
major= df["Major"]
minor=df["Minor"]
good=df["Good"]
other=df["Other"]
name=df["Photo Name"]


### create the train lables (0-good,1-minor,2-major, 3-other, 4-fan)
class_names=[ "good","minor", "major","other"]
num_images=name.size
lables=[]

##TODO: The fan class can be =4, when we will do multi lableing. 
for i in range(num_images):
    lables.insert(i, 3) #by default other.
    if(good[i]==1):
        lables.insert(i,0)
        #copy to good folder
        print("copying.. to good...\n")
        copyfile("./image_data/"+name[i],"./data/good/"+name[i])
    elif(minor[i]==1):
        lables.insert(i,1)
        #copy to  minor folder
        print("copying.. to minor...\n")
        copyfile("./image_data/"+name[i],"./data/minor/"+name[i])
    elif(major[i]==1):
        lables.insert(i,2)
        #copy to major folder
        print("copying.. to major...\n")
        copyfile("./image_data/"+name[i],"./data/major/"+name[i])
    else:
        lables.insert(i, 3)
        #copy to other folder
        print("copying.. to other...\n")
        copyfile("./image_data/"+name[i],"./data/other/"+name[i])
        
print("images labled...")
```
